```python
# Let's start visualizing!

borough_rooms = lsts.groupby(['neighbourhood_group', 'room_type'])['id'].count().reset_index()
borough_rooms
```

```python
sns.set(rc={'figure.figsize': (16, 5)})
ax = sns.barplot(x = 'neighbourhood_group', y = 'id', hue = 'room_type', data = borough_rooms, 
                 palette ='mako', ci = False)
```

```python
listings_borough = lsts.groupby(['neighbourhood_group'])['id'].count().reset_index()
listings_borough
```

```python
plot = listings_borough.plot.pie(y='id', figsize=(6, 6), 
                                 labels=['Bronx', 'Brooklyn', 'Manhattan', 'Queens', 'Staten Island'])
plt.legend(loc='upper left')
plt.title('Listings by Borough')
plt.show()
```

```python
# Donut chart
names=[r'Bronx (2 %)', r'Brooklyn (40 %)', 
r'Manhattan (44 %)', r'Queens (12 %)', r'Staten Island (0.75 %)']
size=[1192,19474,21595,5962,365]
 
# Create a circle for the center of the plot
my_circle=plt.Circle( (0,0), 0.7, color='white')


plt.rcParams['text.color'] = 'black'
fig = plt.figure()
fig.patch.set_facecolor('white')

plt.pie(size, labels=names, colors=['blue','orange','green', 'red', 'purple'], labeldistance=1.15, 
        wedgeprops = { 'linewidth' : 7, 'edgecolor' : 'white' })
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.title('Listings by Borough')

# plt.savefig('donut.png', dpi=100)

plt.show()
```

```python
lsts_gz['price'] = lsts_gz['price'].str.replace(',', '')
lsts_gz['price'] = lsts_gz['price'].str.replace('$','').astype(str).astype(float)
```

```python
avg_prices = lsts_gz.groupby(['neighbourhood_group_cleansed', 'room_type'])['price'].mean().reset_index()
avg_prices
```

```python
sns.set(rc={'figure.figsize': (16, 5)})
ax = sns.barplot(x = 'neighbourhood_group_cleansed', y = 'price', hue = 'room_type', data = avg_prices, 
                 palette ='plasma_r', ci = False)
```
